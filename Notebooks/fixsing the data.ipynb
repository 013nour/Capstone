```python
RiyadhRastCafe['isLoved'] = 0
RiyadhRastCafe['rated1'] = 0
RiyadhRastCafe['rated2'] = 0
RiyadhRastCafe['rated3'] = 0
RiyadhRastCafe['rated4'] = 0
RiyadhRastCafe['rated5'] = 0
```

```python
reviews = pd.read_csv('labeled_reviews.csv')
reviews
```

```python
np.where(RiyadhRastCafe["place_id"] == reviews["place_id"][0])[0]
```

```python
for i in range(len(reviews)):

    try:
        indx = np.where(RiyadhRastCafe["place_id"] == reviews["place_id"][i])[0][0]

        RiyadhRastCafe.loc[indx, 'isLoved'] = reviews.loc[i, 'isLoved']
        RiyadhRastCafe.loc[indx, 'rated1'] = reviews.loc[i, 'rated1']
        RiyadhRastCafe.loc[indx, 'rated2'] = reviews.loc[i, 'rated2']
        RiyadhRastCafe.loc[indx, 'rated3'] = reviews.loc[i, 'rated3']
        RiyadhRastCafe.loc[indx, 'rated4'] = reviews.loc[i, 'rated4']
        RiyadhRastCafe.loc[indx, 'rated5'] = reviews.loc[i, 'rated5']
        
    except:
        pass

RiyadhRastCafe
```
