```python
import googlemaps
```

```python
gmaps = googlemaps.Client('AIzaSyDIPKmM5qX5wihzclc2ObruIzcUY8GZeUQ')
PriceLevel = {0: 'Free', 1: 'Inexpensive', 2: 'Moderate', 3: 'Expensive', 4: 'Very Expensive'}
```

```python
def addPlaces(lat, lng, KeyWord, df):
    
    location = (lat, lng)

    #best redias is 1500 as the it max return 20 results each respons
    responce = gmaps.places_nearby(location=location, radius=1500, keyword=KeyWord)

    for i in range(len(responce['results'])):
        
        #get place id 
        place_id = responce['results'][i]['place_id']
        
        #get Name
        name = responce['results'][i]['name']
        
        #get restaurant type 
        types = KeyWord
        
        #get Price Level 
        try:
            price_level = responce['results'][i]['price_level']
        except : 
            price_level = 'Null'
            
        #get rating 
        rating = responce['results'][i]['rating']
    
        #get total number of users who rated the restaurant
        user_ratings_total = responce['results'][i]['user_ratings_total']
        
        #get neighborhood info
        try:
            compound_code = responce['results'][i]['plus_code'].get('compound_code')
        except :
            compound_code = 'Null'
        try:
            global_code = responce['results'][i]['plus_code'].get('global_code')
        except : 
            global_code = 'Null'

        #get area near or surrounding 
        vicinity = responce['results'][i]['vicinity']
        
        # get location geo
        lat = responce['results'][i]['geometry'].get('location').get('lat')
        lng = responce['results'][i]['geometry'].get('location').get('lng')
        
        df.loc[len(df.index)] = [place_id, name, types, price_level, rating, user_ratings_total, compound_code, global_code, lat, lng]
```

```python
RiyadhRastCafe = pd.DataFrame(columns=['place_id', 'name', 'types', 'price_level', 'rating', 
                                       'user_ratings_total', 'compound_code', 'global_code', 'lat', 'lng'])
keyword='Restaurants'
for i in tqdm(range(len(Riyadh))):
    addPlaces(lat=Riyadh['lat'][i], lng=Riyadh['lng'][i], KeyWord=keyword, df=RiyadhRastCafe) 
```

# Get Reviews

```python
def getReview(place_id):
    
    place_result = gmaps.place(place_id=place_id,language='ar')
    indx = np.where(RastCafe["place_id"] == place_id)[0][0]
    
    #name 
    RastCafe.loc[indx,'name'] = place_result['result']['name']
    #place 
    RastCafe.loc[indx,'place'] = place_result['result']['address_components'][1]['long_name']
    #place neighbhood 
    RastCafe.loc[indx,'neighbhood'] =  place_result['result']['address_components'][2]['long_name']
    #place city 
    RastCafe.loc[indx,'city'] =  place_result['result']['address_components'][3]['long_name']
    #place provinces 
    try:
        RastCafe.loc[indx,'provinces'] = place_result['result']['address_components'][4]['long_name']
    except:
        RastCafe.loc[indx,'provinces'] = 'Null'

    #num of opeining days 
    try :
        RastCafe.loc[indx,'opeining_days'] = len(place_result['result']['current_opening_hours']['weekday_text'])
    except:
        RastCafe.loc[indx,'opeining_days'] = 'Null' 
    
    #delivery
    try:
        RastCafe.loc[indx,'delivery'] = place_result['result']['delivery']
    except:
        RastCafe.loc[indx,'delivery'] = False
        
    #dine_in
    try: 
        RastCafe.loc[indx,'dine_in'] = place_result['result']['dine_in']
    except: 
        RastCafe.loc[indx,'dine_in'] = False

    #reviews
    try:
        for j in range(len(place_result['result']['reviews'])):
            rating = place_result['result']['reviews'][j]['rating']
            review = place_result['result']['reviews'][j]['text']
    
            reviews.loc[len(reviews.index)] = [place_id, rating, review]
    except:
        pass
    
    #photos 
    try:
        for k in range(len(place_result['result']['photos'])):
            ref = place_result['result']['photos'][k]['photo_reference']
            photos.loc[len(photos.index)] = [place_id, ref]
    except:
        pass
```

# Find number of utilities surrounding the place. Include : officess, companies, schools, hotels, Parks

```python
def addUtilities(lat, lng, place_id):
    
    location = (lat, lng)
    indx = np.where(RastCafe["place_id"] == place_id)[0][0]

    #best redias is 1500 as the it max return 20 results each respons
    try:
        responce = gmaps.places_nearby(location=location, radius=1500, keyword='School')
        RastCafe.loc[indx,'schoolNum'] = len(responce['results'])
    except:
        RastCafe.loc[indx,'schoolNum'] = 0
    
    try:
        responce = gmaps.places_nearby(location=location, radius=1500, keyword='office')
        RastCafe.loc[indx,'officeNum'] = len(responce['results'])
    except:
        RastCafe.loc[indx,'officeNum'] = 0

    try: 
        responce = gmaps.places_nearby(location=location, radius=1500, keyword='company')
        RastCafe.loc[indx,'companieNum'] = len(responce['results'])
    except:
        RastCafe.loc[indx,'companieNum'] = 0
        
    try:
        responce = gmaps.places_nearby(location=location, radius=1500, keyword='hotel')
        RastCafe.loc[indx,'HotelNum'] = len(responce['results'])
    except:
        RastCafe.loc[indx,'HotelNum'] = 0 
        
    try:
        responce = gmaps.places_nearby(location=location, radius=1500, keyword='park')
        RastCafe.loc[indx,'parksNum'] = len(responce['results'])
    except:
        RastCafe.loc[indx,'parksNum'] = 0 
```

```python
RastCafe['schoolNum'], RastCafe['companieNum'] = np.NaN, np.NaN
RastCafe['officeNum'], RastCafe['HotelNum'] = np.NaN, np.NaN
```

```python
for i in tqdm(range(len(RastCafe))):
    addUtilities(lat= RastCafe['lat'][i], lng= RastCafe['lng'][i], place_id= RastCafe['place_id'][i]) 
```

# Find nearby rent prices 

```python
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
import re
```

```python
def getStors():
    for i in tqdm(range (1, 21)):
        
        aqar_path = 'https://sa.aqar.fm/%D9%85%D8%AD%D9%84%D8%A7%D8%AA-%D9%84%D9%84%D8%A5%D9%8A%D8%AC%D8%A7%D8%B1/%D8%A7%D9%84%D8%B1%D9%8A%D8%A7%D8%B6/'+str(i)
    
        driver = webdriver.Edge()
        driver.get(aqar_path)
        source = driver.page_source
        driver.quit()
        soup = BeautifulSoup(source, 'html.parser')

        name = soup.find_all(class_='_footer__CnldH')
        price = soup.find_all(class_='_price__X51mi')
        spc = soup.find_all(class_='_spec__SIJiK')

        for i in range(len(name)):
            places.append(name[i].text)
            prices.append(price[i].text) 
            space.append(spc[i].text)
              
        # get the page for each store to get the location geo later on 
        for a in soup.find_all('a', href=True):
            m = re.search(r'\d+$', a['href'])
            if m is not None and len(m.group()) > 5 and ("https:" in a['href']) == False:
                urls.append('https://sa.aqar.fm'+a['href'])
```
